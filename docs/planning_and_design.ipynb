## Feature Specification: User Authentication & Management

### 1. User Stories (The "Why")

#### For the System Administrator:

*   **Story 1: Onboarding New Employees**
    > **As an** administrator,
    > **I want to** create new user accounts with their name, email, and role (Admin/User),

#### For the Standard User (Salesperson, Account Manager, etc.):

*   **Story 2: Accessing the System**
    > **As a** user,
    > **I want to** log in with my email and password,
    > **so that** I can securely access the CRM to perform my daily tasks.

*   **Story 3: Maintaining a Session**
    > **As a** user,
    > **I want to** remain logged in as I navigate the application and even if I close the browser tab,
    > **so that** I don't have to waste time re-entering my credentials repeatedly during my workday.

*   **Story 4: Securely Exiting the System**
    > **As a** user,
    > **I want to** explicitly log out of the application,
    > **so that** I can ensure my session is securely terminated, especially when using a shared computer.

### 2. Technical Specification (The "How")

This section details the implementation required to satisfy all the user stories above.

#### **Data Model: `users` Table**
*   **Fields:**
    *   `id` (Primary Key)
    *   `name` (String)
    *   `email` (String, must be unique)
    *   `password_hash` (String) - **CRITICAL:** Never store plain-text passwords. Use a strong, one-way hashing algorithm like **bcrypt**.
    *   `role` (Enum: 'user', 'admin')
    *   `is_active` (Boolean, default true)

#### **User Creation (Admin-Only)**
*   **User Action:** An authenticated Admin navigates to a "User Management" section in the application settings.
*   **UI Flow:**
    *   Admin clicks a "New User" button.
    *   A form appears to enter the new user's `Name`, `Email`, and select their `Role`.
*   **Backend Interaction:**
    *   **Endpoint:** `POST /api/admin/users` (This route must be protected and only accessible by users with the 'admin' role).
    *   **Logic:**
        1.  Validate the request data (e.g., is the email format valid? Is the email already in use?).
        2.  Generate a secure, temporary password.
        3.  **Hash the temporary password** using bcrypt.
        4.  Save the new user record to the `users` table.
        5.  **Best Practice:** Trigger an email to the new user's address containing their login details and a link to a page where they must set their own permanent password.
*   **User Feedback:** A success message is shown to the Admin: "User '[New User Name]' created and welcome email sent."

#### **Login Process**
*   **User Action:** A non-authenticated user accesses the application and is presented with the login page. They enter their email and password and click "Login".
*   **UI Flow:** A simple form with "Email" and "Password" fields and a "Login" button.
*   **Backend Interaction:**
    *   **Endpoint:** `POST /api/auth/login`
    *   **Logic:**
        1.  Find the user in the database by their email address.
        2.  If no user is found, return a generic error. **Do not** reveal that the email address doesn't exist.
        3.  If a user is found, use the bcrypt library to compare the submitted password with the `password_hash` stored in the database.
        4.  If the passwords match, generate a session token (**JWT - JSON Web Token** is the recommended standard).
        5.  Return the JWT to the client.
*   **User Feedback:**
    *   **On Success:** The user is redirected to the application's Home Page (`/home`). The JWT is stored securely on the client (e.g., in an `httpOnly` cookie).
    *   **On Failure:** The login page displays a single, generic error message: "Invalid email or password."

#### **Session Management (Authenticated Requests)**
*   **Frontend Logic:** For every subsequent API request to a protected endpoint (e.g., `GET /api/accounts`), the browser must automatically send the stored JWT (in the cookie) with the request.
*   **Backend Logic:** A **middleware** layer must be applied to all protected API routes. This middleware will:
    1.  Inspect the request for the JWT.
    2.  Validate the JWT's signature and expiration date.
    3.  If the token is valid, allow the request to proceed to the intended controller.
    4.  If the token is invalid, expired, or missing, reject the request with a `401 Unauthorized` status code.

#### **Logout Process**
*   **User Action:** An authenticated user clicks the "Logout" button, typically in a user profile dropdown menu.
*   **UI Flow:** The application immediately invalidates the user's session on the client side.
*   **Backend Interaction:**
    *   **Endpoint:** `POST /api/auth/logout`
    *   **Logic:** The client-side application should simply delete the stored JWT. The backend can optionally implement a token "blacklist" for higher security, but for an internal tool, client-side deletion is often sufficient.
*   **User Feedback:** The user is immediately redirected to the login page.

### 2. Technical Specification (The "How")

#### **Data Model: `leads` Table**
*   **Fields:** `id`, `first_name`, `last_name`, `company` (required), `email`,`lead_source`, `phone`, `status` (Enum: 'New', 'Contacted', 'Qualified', 'Converted'), `owner_id` (FK to `users.id`).

#### **CREATE**
*   **User Action:** Clicks "New Lead" button.
*   **UI Flow:** A modal form appears for Lead details. `Last Name`, `Company`, and `Status` are required.
*   **Backend Interaction:** `POST /api/leads`. Validates data, creates the record, assigns the current user as `owner_id`.
*   **User Feedback:** On success, redirects to the new Lead's detail page with a success toast.

*   **List View:**
    *   **User Action:** Navigates to "/leads".
    *   **UI Flow:** Displays a table of all Leads where `status != 'Converted'`.
    *   **Backend Interaction:** `GET /api/leads`. Supports filtering by `status` and `owner_id`.
*   **Detail View:**
    *   **User Action:** Clicks a Lead's name.
    *   **UI Flow:** Navigates to `/leads/{id}`. Displays all Lead fields and a related list for Activities.
    *   **Backend Interaction:** `GET /api/leads/{id}`.

#### **UPDATE**
*   **User Action:** Clicks "Edit" on the detail page. The most common update is changing the `Status` field.
*   **UI Flow:** Fields become editable.
*   **Backend Interaction:** `PUT /api/leads/{id}`.
*   **User Feedback:** "Save successful" toast.

#### **BUSINESS LOGIC: Lead Conversion**
*   **User Action:** Clicks the "**Convert**" button on a qualified Lead's detail page.
*   **UI Flow:** A dedicated conversion modal appears. It shows the data that will be used to create the new records and requires the user to enter a name for the new **Opportunity**.
*   **Backend Interaction:**
    *   **Endpoint:** `POST /api/leads/{id}/convert`
    *   **Logic:** This is a critical transaction.
        1.  **Start a database transaction.**
        2.  Create a new `Account` record using `company` name from the Lead.
        3.  Create a new `Contact` record using name/email/phone from the Lead, linking it to the new Account's ID.
        4.  Create a new `Opportunity` record with the user-provided name, linking it to the new Account's ID.
        5.  Update the original `Lead`'s status to 'Converted'.
        6.  **Commit the transaction.** If any step fails, the entire operation must be rolled back.
*   **User Feedback:** A success screen appears, confirming the conversion and providing direct links to the newly created Account, Contact, and Opportunity records.

## Feature Specification: Activity Management (Tasks & Events)

### 1. User Stories (The "Why")


*   **Story 1: Creating a Contextual To-Do Item**
    > **As a** salesperson,
    > **I want to** create a Task like "Follow up on proposal" and relate it *both* to the 'Awesome Project' Opportunity *and* the 'Jane Doe' Contact,
    > **so that** the task appears in the timeline for the deal and also in my activity history with that specific person.

*   **Story 2: Scheduling a Client Meeting**
    > **As an** account manager,
    > **I want to** schedule an Event (a meeting) and link it to the client's Account record,
    > **so that** anyone viewing the Account can see that a meeting is scheduled.

*   **Story 3: Viewing My Daily Agenda**
    > **As a** user,
    > **I want to** see a single, unified list of all my open Tasks and upcoming Events, sorted chronologically,
    > **so that** I have a clear picture of my workload and schedule for the day.


### 2. Technical Specification (The "How")

#### **Data Model: `activities` Table**
*   **Fields:**
    *   `id` (PK)
    *   `type` (Enum: 'Task', 'Event', required)
    *   `subject` (String, required)
    *   `description` (Text, optional)
    *   `status` (Enum: 'Not Started', 'In Progress', 'Completed', 'Waiting on someone else', 'Deferred')
    *   `priority` (Enum: 'High', 'Normal', 'Low', for Tasks)
    *   `due_date` (Date, for Tasks)
    *   `start_time` (DateTime, for Events)
    *   `end_time` (DateTime, for Events)
    *   `is_all_day_event` (Boolean, default false, for Events)
    *   `location` (String, optional, for Events)
    *   `assigned_to_id` (FK to `users.id`, required)
    *   `what_id` (Integer, nullable) - ID of the non-person object.
    *   `what_type` (String, nullable) - Model name: 'Account', 'Opportunity', 'Case'.
    *   `who_id` (Integer, nullable) - ID of the person object.
    *   `who_type` (String, nullable) - Model name: 'Contact', 'Lead'.

#### **CREATE**
*   **User Action:** Clicks "New Task" or "New Event" from a global button or a related list.
*   **UI Flow:**
    *   A modal form appears, titled either "New Task" or "New Event".
    *   The form dynamically displays the correct fields based on the type:
        *   **If 'Task':** Shows `Due Date`, `Status`, `Priority`.
        *   **If 'Event':** Shows `Start/End Time`, `Location`, `All-Day Event` checkbox.
    *   Both forms include `Subject`, `Assigned To`, `Related To` (What), and `Name` (Who).
*   **Backend Interaction:**
    *   **Endpoint:** `POST /api/activities`
    *   **Request Body:** A JSON object containing all relevant details. The `type` field is mandatory.
*   **User Feedback:** The new activity appears in relevant lists/views. A success toast is shown.

#### **READ**
*   **Task View ("To-Do List"):**
    *   **UI Flow:** A dedicated page showing a list of all activities where `type = 'Task'` and `status != 'Completed'`, assigned to the current user.
    *   **Backend Interaction:** `GET /api/activities?type=Task&assigned_to=me&status_ne=Completed`
*   **Event View ("Calendar"):**
    *   **UI Flow:** A dedicated page showing a calendar view (e.g., using a library like FullCalendar).
    *   **Backend Interaction:** `GET /api/activities?type=Event&assigned_to=me`. The API should accept date range parameters (e.g., `start_date`, `end_date`) to fetch events for the visible month/week.
*   **Contextual View (Activity Timeline):**
    *   **UI Flow:** Appears on detail pages for Accounts, Opportunities, etc.
    *   **Backend Interaction:** The API for the parent record (e.g., `GET /api/opportunities/{id}`) should return a sorted list of all related activities.



## Non-Functional Requirements (NFRs)

These requirements define the quality attributes and operational standards of the CRM system. They are critical for ensuring the application is reliable, secure, and fit for its purpose as a core business tool for the web agency.

### 1. Reliability & Durability

*   **Statement:** The system must ensure high data durability and reliability. Core business data (Accounts, Contacts, Leads, etc.) must be persistently stored and protected against loss.
*   **Context & Quantification:**
    *   Losing client or project data would be significantly detrimental to the agency's operations and reputation.
    *   The system **shall** utilize a database with transactional integrity (ACID-compliant), such as PostgreSQL.
    *   Regular, automated backups (daily) **shall** be implemented with a **Recovery Point Objective (RPO) of less than 24 hours** (meaning no more than one day's worth of data could be lost in a worst-case scenario).
    *   The system **shall** have a **Recovery Time Objective (RTO) of less than 4 hours** during business hours, meaning the system can be restored from a backup within that timeframe.

### 2. Security

*   **Statement:** The system must provide robust security to protect sensitive client and agency data from unauthorized access or modification.
*   **Context & Quantification:**
    *   All user access **shall** be protected by secure authentication, including individual user accounts and strongly hashed passwords (e.g., using bcrypt).
    *   Data transmission between the client (browser) and server **shall** be encrypted using HTTPS (SSL/TLS).
    *   The system **shall** implement Role-Based Access Control (RBAC), at a minimum distinguishing between 'Admin' and 'User' roles to control access to sensitive functions like user creation.
    *   The application **shall** be built following secure coding practices to mitigate common web vulnerabilities (OWASP Top 10). This includes using framework features for protection against SQL Injection and XSS, and validating all incoming data.
    *   A process **shall** be in place for regularly scanning and updating third-party libraries and dependencies to patch known vulnerabilities.

### 3. Performance & Responsiveness

*   **Statement:** The system must provide a responsive and fluid user experience for all common operations under the expected load of a small team.
*   **Context & Quantification:**
    *   For a small number of concurrent users (1-5), the following performance targets **should** be met:
        *   Loading core list views (Accounts, Contacts, etc.) with up to a few thousand records: **< 3 seconds**.
        *   Loading individual record detail views: **< 2 seconds**.
        *   Saving or updating a record (from click to user feedback): **< 1.5 seconds**.
    *   These performance targets **should** be met through efficient database queries (e.g., proper indexing of foreign keys and frequently filtered columns) rather than complex caching mechanisms for the MVP. Performance profiling will be a post-launch activity if targets are consistently missed.

### 4. Maintainability

*   **Statement:** The system must be designed and built in a way that facilitates ongoing maintenance, bug fixing, and future feature development by a solo developer or small team.
*   **Context & Quantification:**
    *   The codebase **shall** adhere to consistent style guides (e.g., PEP 8 for Python, Prettier for JavaScript) and include clear documentation for complex business logic.
    *   The architecture **shall** have a clean separation of concerns (frontend React application, backend Django API, PostgreSQL database).
    *   The system **shall** be structured to facilitate automated testing (e.g., unit tests for business logic, integration tests for API endpoints).
    *   All code and configuration **shall** be managed under version control (Git) with a consistent branching strategy.

### 5. Operability

*   **Statement:** The system must be operable, providing sufficient visibility into its health and behavior to allow for effective troubleshooting and production management.
*   **Context & Quantification:**
    *   To enable effective debugging, the backend application **shall** produce structured logs (e.g., JSON format) for key events, especially for errors and handled exceptions.
    *   The backend **shall** expose a simple `/api/health` endpoint that can be used to confirm the service is running and can connect to the database.
    *   All environment-specific configuration (database credentials, secret keys, etc.) **shall** be managed via environment variables and **must not** be hardcoded in the source code.
